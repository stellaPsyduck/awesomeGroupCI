```python
# Plotting data for our pre-processing section
import pandas as pd
import os
from plotnine import ggplot, aes, geom_line, labs, scale_x_datetime
import numpy as np

ticker = "ACIW"
dataset = pd.read_csv(f'{ticker}_unadjusted_prices_2007_2018.csv', skiprows=2)
dataset.rename(columns={"Unnamed: 1": "Price"}, inplace=True)
dataset["Price"] = np.log(dataset["Price"]) # I doubt anything is 0....

dataset['Year'] = pd.to_datetime(dataset['Date']).dt.year

(ggplot(dataset) +
         aes(x='Date', y='Price', group = 1) +
         scale_x_datetime(date_breaks='1 year', date_labels='%Y')+
         labs(title=f'{ticker} Scaled Data',       
           x='Date', y='Stock Price (USD)') +
         geom_line())
```
